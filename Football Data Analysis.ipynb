```python
#pie chart of penalty scored vs missed

plt.figure(figsize=(13,6))

penalty_missed=df.Penalty_Attempted.sum()-total_goals
data=[penalty_missed,total_goals]
label=['Penalty_Missed','Penalty_Scored']
color=sns.color_palette('Set2')
plt.pie(data,labels=label,colors=color,autopct="%0.0f%%")
plt.show()
```

```python
df['Club'].value_counts().nlargest(5).plot(kind='bar',figsize=(14,7),color=sns.color_palette('Paired'))
```

Clubs with maximum number of players

```python
#Clubs with least number of players


df['Club'].value_counts().nsmallest(5).sort_values(ascending=False).plot(kind='bar',figsize=(14,7),color=sns.color_palette('tab10'))
```

```python
#players based on age group

under20=df[df['Age']<=20]
age20_25=df[(df['Age']>20) & (df['Age']<=25)]
age25_30=df[(df['Age']>25) & (df['Age']<=30)]
above30=df[df['Age']>30]
```

```python
x=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
label=['<20','>20 & <=25','>25 & <=30','>30']
plt.figure(figsize=(13,6))
plt.title("Players Age Group",fontsize=30)
plt.pie(x,labels=label, colors=sns.color_palette('tab10'),autopct="%0.1f%%")
plt.show()
```

```python
#Total under 20 players
under20_each_club=df[df['Age']<20]
under20_each_club['Club'].value_counts().plot(kind='bar',figsize=(12,6),color=sns.color_palette('rocket'))
```

```python
#Under 20 Players in Manchester United

under20_each_club[under20_each_club['Club']=='Manchester United']
```

```python
# Under 20 players in chelsea

under20_each_club[under20_each_club['Club']=='Chelsea']
```

```python
# Average age of players in each club
plt.figure(figsize=(14,7))
sns.boxplot(x='Club',y='Age', data=df)
plt.xticks(rotation=90)
```

```python
num_player=df.groupby("Club").size()
avg_age=df.groupby("Club")['Age'].sum()/num_player
avg_age.sort_values(ascending=False)
```

```python
#total assists from each club

assist_club=pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=assist_club.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assist",fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title("Plot of Club vs Total Assists",fontsize=30)
```

```python
# top 10 assists

top_assists=df[['Name','Club','Assists','Matches']].nlargest(n=10, columns='Assists')
top_assists
```

```python
# Goals by Clubs

goals_by_clubs=pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=goals_by_clubs.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=20)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation=75)
plt.title("Goals Scored by Clubs",fontsize=30)
plt.rcParams['figure.figsize']=(20,8)
```
